```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
```

```python
gt_by_city = pd.read_csv('~/global_weather_Python/sv_files/GlobalLandTemperaturesByCity.csv')
gt_by_country = pd.read_csv('~/global_weather_Python/sv_files/GlobalLandTemperaturesByCountry.csv')
gt_by_major_city = pd.read_csv('~/global_weather_Python/sv_files/GlobalLandTemperaturesByMajorCity.csv')
gt_by_state = pd.read_csv('~/global_weather_Python/sv_files/GlobalLandTemperaturesByState.csv')
gt = pd.read_csv('~/global_weather_Python/sv_files/GlobalTemperatures.csv')
```

```python
ust = gt_by_city[gt_by_city.Country == 'United States']
```

```python
la = ust[ust.City == 'Los Angeles']
la = la[la.dt == '2013-04-01']
philly = ust[ust.City == 'Philadelphia']
philly = philly[philly.dt == '2013-04-01']
```

```python
la_philly = pd.concat([la, philly])
```

```python
float(re.findall(r'\d+.\d+', str(philly.Latitude))[1])
```

```python
m = Basemap(width=12000000, height=9000000, projection='lcc', resolution='c', lat_0=50, lat_1=45, lat_2=55, 
            lon_0=-105)
```

```python
m.drawcoastlines()
m.drawmapboundary()
m.drawcountries()
lon, lat = -float(re.findall(r'\d+.\d+', str(philly.Longitude))[1]), float(re.findall(r'\d+.\d+', str(philly.Latitude))[1])
xtp, ytp = m(lon, lat)
lonpt, latpt = m(xtp, ytp, inverse=True)
m.plot(xtp, ytp, 'bo')
plt.show()
```
